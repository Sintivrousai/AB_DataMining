```python
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame
data = {
    "Maintenance": [0.182624, 0.165925, 0.124343],
    "Routine": [0.328064, 0.338750, 0.337071],
    "Stock-Up": [0.141073, 0.099686, 0.045857],
    "Top-Up": [0.348239, 0.395639, 0.492730],
}

df = pd.DataFrame(data)

# Generate pie charts for each row
for idx, row in df.iterrows():
    labels = row.index  # Column names
    sizes = row.values  # Percentage values

    # Create the pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(
        sizes,
        labels=labels,
        autopct="%1.1f%%",
        startangle=90,  # Start from 90 degrees
    )

    plt.title(f"Cluster {idx}")
    plt.show()
```
